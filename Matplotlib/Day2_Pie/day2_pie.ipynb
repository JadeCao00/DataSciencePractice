1. 绘制饼图(类别最好不要超过六类)

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams # 字体
rcParams['font.family'] = 'SimHei' # SimHei中文字体

# 创建图表，设置大小
plt.figure(figsize=(10, 5))

# 数据
things_to_do = ['学习', '运动', '娱乐', '睡觉', '其他']
time = [6, 1, 3, 9, 5]
colors=['#66b3ff', '#99ff99', '#ffcc99', '#ff9999', '#ff4499'] # 配色

# 绘制饼图
plt.pie(time, labels=things_to_do,
       autopct='%1.1f%%', # 显示各个部分占比数据
       startangle=90, # 调整初始画图角度
       colors = colors) # 设置配色

# 添加标题
plt.title('一日时间分布', color='red', fontsize=20)

# 自动优化排版
plt.tight_layout()

# 显示图标
plt.show()
```

2. 绘制环形图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams # 字体
rcParams['font.family'] = 'SimHei' # SimHei中文字体

# 创建图表，设置大小
plt.figure(figsize=(10, 5))

# 数据
things_to_do = ['学习', '运动', '娱乐', '睡觉', '其他']
time = [6, 1, 3, 9, 5]
colors=['#66b3ff', '#99ff99', '#ffcc99', '#ff9999', '#ff4499'] # 配色

# 绘制环形图
plt.pie(time, labels=things_to_do,
       autopct='%1.1f%%', # 显示各个部分占比数据
       startangle=90, # 调整初始画图角度
       colors = colors,
       wedgeprops={'width':0.6}, # 设置圆环宽度百分比
        pctdistance=0.7, # 设置百分比数字的位置
       ) 

# 添加标题
plt.title('一日时间分布', color='red', fontsize=20)
plt.text(0,0, '总计:\n100%', ha='center')

# 自动优化排版
plt.tight_layout()

# 显示图标
plt.show()
```

3. 绘制爆炸式饼图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams # 字体
rcParams['font.family'] = 'SimHei' # SimHei中文字体

# 创建图表，设置大小
plt.figure(figsize=(10, 5))

# 数据
things_to_do = ['学习', '运动', '娱乐', '睡觉', '其他']
time = [6, 1, 3, 9, 5]
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff9999', '#ff4499'] # 配色
explode = [0.1, 0, 0, 0, 0] # 设置突出块的位置

# 绘制饼图
plt.pie(time, labels=things_to_do,
       autopct = '%1.1f%%', # 显示各个部分占比数据
       startangle = 0, # 调整初始画图角度
       colors = colors, # 设置配色
       explode = explode, # 设置突出块
        shadow=True,
       ) 
# 添加标题
plt.title('一日时间分布', color='red', fontsize=20)

# 自动优化排版
plt.tight_layout()

# 显示图标
plt.show()
```
