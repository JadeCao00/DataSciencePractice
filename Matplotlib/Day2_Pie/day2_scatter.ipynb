```python
import matplotlib.pyplot as plt
import random
from matplotlib import rcParams # 字体
rcParams['font.family'] = 'SimHei' # SimHei中文字体

# 创建图表，设置大小
plt.figure(figsize=(10, 8))

# 数据
x = []
y = []
for i in range(800):
    tmp = random.uniform(0, 10)
    x.append(tmp) 
    y.append(2*tmp + random.gauss(0,2))


# 绘制散点图
plt.scatter(x, y,
           color='blue',
           alpha=0.8,
           s = 20, # 圆点大小
           label='数据'
           )

# 添加标题
plt.title('X变量与Y变量的关系', color='red', fontsize=20)

# 添加坐标轴的标签
plt.xlabel('X自变量', fontsize=15)
plt.ylabel('Y因变量', fontsize=15)

# 添加图例
plt.legend(loc='upper left')

# 添加网格线
plt.grid(True, alpha=0.1, color='blue', linestyle='--') # alpha -透明度, --虚线, -实线
# plt.grid(axis='y')

# 设置刻度字体大小
plt.xticks(rotation=0, fontsize=12)
plt.yticks(rotation=0, fontsize=12)

# 设置y轴范围
plt.ylim(0, 30)

plt.plot([0, 10], [0, 20], color='red', linewidth=2) # 正常情况下，折线图的X值范围和Y值范围需要计算

# 显示图标
plt.show()
```
