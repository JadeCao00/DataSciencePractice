1. 绘制多个图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams # 字体
rcParams['font.family'] = 'SimHei' # SimHei中文字体

# 数据
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [100, 150, 80, 130, 120, 90]

graph1 = plt.subplot(2, 2, 1) # 生成一个子图 ()内是几行几列第几个图（索引）
graph1.plot(month, sales)

graph2 = plt.subplot(2, 2, 2)
graph2.bar(month, sales)
```
