1. 绘制折线图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams # 字体
rcParams['font.family'] = 'SimHei' # SimHei中文字体

# 创建图表，设置大小
plt.figure(figsize=(10, 5))

# 数据
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [100, 150, 80, 130, 120, 90]

# 绘制折线图
plt.plot(month, sales,
        label='产品A',
        color='orange',
        linewidth=2,
        linestyle='-',
        marker='o',)

# 添加标题
plt.title('2025年销售趋势', color='red', fontsize=20)

# 添加坐标轴的标签
plt.xlabel('月份', fontsize=15)
plt.ylabel('销售额（万元）', fontsize=15)

# 添加图例
plt.legend(loc='upper left')

# 添加网格线
plt.grid(True, alpha=0.1, color='blue', linestyle='--') # alpha -透明度, --虚线, -实线
# plt.grid(axis='y')

# 设置刻度字体大小
plt.xticks(rotation=0, fontsize=12)
plt.yticks(rotation=0, fontsize=12)

# 设置y轴范围
plt.ylim(0, 160)

# 在每个数据点上显示数值
for x,y in zip(month, sales):
    plt.text(x, y+2, str(y), ha='center', va='bottom', fontsize=10)

# 显示图标
plt.show()
```
