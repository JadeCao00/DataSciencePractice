1. 绘制柱状图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams # 字体
rcParams['font.family'] = 'SimHei' # SimHei中文字体

# 创建图表，设置大小
plt.figure(figsize=(10, 5))

# 数据
subjects = ['语文', '数学', '英语', '生物', '地理', '化学']
scores = [110, 120, 130, 80, 75, 70]

# 绘制柱状图
plt.bar (subjects,scores,
        label='小明',
        color='orange',
        width=0.6)

# 添加标题
plt.title('小明2025年期末成绩分布', color='red', fontsize=20)

# 添加坐标轴的标签
plt.xlabel('科目', fontsize=15)
plt.ylabel('分数', fontsize=15)

# 添加图例
plt.legend(loc='upper right')

# 添加网格线
plt.grid(axis='y', alpha=0.1, color='blue', linestyle='-') # alpha -透明度, --虚线, -实线
# plt.grid(axis='y')

# 设置刻度字体大小
plt.xticks(rotation=0, fontsize=12)
plt.yticks(rotation=0, fontsize=12)

# 设置y轴范围
plt.ylim(0, 150)

# 在每个数据点上显示数值
for x,y in zip(subjects, scores):
    plt.text(x, y+2, str(y), ha='center', va='bottom', fontsize=10)

# 自动优化排版
plt.tight_layout()

# 显示图标
plt.show()
```

2. 绘制条形图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams # 字体
rcParams['font.family'] = 'SimHei' # SimHei中文字体

# 创建图表，设置大小
plt.figure(figsize=(10, 5))

# 数据
countries = ['United States', 'China', 'Japan', 'United Kingdom', 'India', 'Australia']
gdp = [200, 150, 70, 100, 50, 40] # 随便写的

# 绘制条形图
plt.barh (countries,gdp,
        color='blue',
         label='国家')

# 添加标题
plt.title('2025年GDP排名', color='red', fontsize=20)

# 添加坐标轴的标签
plt.xlabel('GDP', fontsize=15)
plt.ylabel('国家', fontsize=15)

# 添加图例
plt.legend(loc='upper right')

# 添加网格线
plt.grid(axis='y', alpha=0.1, color='blue', linestyle='-') # alpha -透明度, --虚线, -实线
# plt.grid(axis='y')

# 设置刻度字体大小
plt.xticks(rotation=0, fontsize=12)
plt.yticks(rotation=0, fontsize=12)

# 设置x轴范围
plt.xlim(0, 200)

# 在每个数据点上显示数值
for x,y in zip(countries, gdp):
    plt.text(y+4,x, str(y), ha='center', va='bottom', fontsize=10)

# 自动优化排版
plt.tight_layout()

# 显示图标
plt.show()
```
